```python
# 라이브러리 불러오기
import seaborn as sns

# titanic 데이터셋 가져오기
df = sns.load_dataset('titanic')

# isnull() 메서드를 합계로 집계하여 각 열의 NaN 개수 계산하기
df.isnull().sum(axis=0) # isnull -> 결측치(NaN)가 있는지 확인하는 메서드 isna와 동일
```

```python
df.iloc[:,:2].sum(axis=0)
```

```python
# NaN 값이 500개 이상인 열을 모두 삭제 - deck 열(891개 중 688개의 NaN 값)
df_thresh = df.dropna(axis=1, thresh=500)  # 결측치 값이 500개 이상이면 삭제하기떄문에 결측치 개수를 알아야함.
print(df_thresh.columns)
```

```python
# age 열에 나이 데이터가 없는 모든 행을 삭제 - age 열(891개 중 177개의 NaN 값)
df_age = df.dropna(subset=['age'], how='any', axis=0)  
print(len(df_age))
```

```python
# 데이터 프레임에서 삭제가 일어나면 인덱스가 사퀸스하지 않다.
df_age.reset_index() # index를 재정의해주는거임.
df_age.reset_index(drop=True) # 기존 인덱스를 새 열로 보존하지 않고 버리고 재정의.
```

```python
# age 열, deck 열 양쪽 모두 데이터가 없는 행을 삭제 
df_age_deck = df.dropna(subset=['age', 'deck'], how='all', axis=0)  
print(len(df_age_deck))
```
