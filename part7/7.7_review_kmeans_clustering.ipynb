#### k-mean K-평균 -  비지도 학습  군집(Cluster) / 여기서 핵심은 K 임.

'''
초기 중심점을 임의로 설정을 합니다. (Centroid)

데이터를 할당합니다. (Assignment)

중심점을 업데이트합니다. (Update)
'''

```python
# 6명 학생 수학 영어점수 2개의 그룹을 나눠보자

import matplotlib.pyplot as plt
import numpy as np
X = np.array([ # Assignment 데이터를 할당 줌
    [2,1] , [2,6], [4,3],
    [5,5], [5,6], [2,3]
])

plt.scatter(X[:,0], X[:,1],s=100,color='yellow',label='Data Point')
plt.grid(True)
plt.show()
```

```python
# 중심점을 할당해보자

centroid_inital = np.array( [[1,1] ,  [5,4]]  )

plt.scatter(X[:,0], X[:,1],s=100,color='yellow',label='Data Point')
plt.scatter(centroid_inital[:,0], centroid_inital[:,1],s=100,color=['red','blue'], marker='X',label='initial centroid')
plt.grid(True)

```

```python
# 데이터 할당 : 모든 데이터 포인트를 순회하면서 두 중심점 중에 더 가까운 클러스터로 할당
# 빨간 클러스터에 할당 되는것은 P1 , P2 , P3
# 파란 클러스터에 할당 되는것은 P4 , P5 , P6
# 각 점이 어느 중심점에서 가까운지 계산하는 것.
```

```python
label_initial = np.array([0,0,0,1,1,1])
plt.scatter(X[label_initial == 0, 0],X[label_initial == 0, 1],s=100,color='lightcoral', label='cluster1')
plt.scatter(X[label_initial == 1, 0],X[label_initial == 1, 1],s=100,color='lightblue', label='cluster2')

plt.scatter(centroid_inital[:,0], centroid_inital[:,1],s=100,color=['red','blue'], marker='X', label='initial centroid')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 중심점 업데이트

X[label_initial==0]
c1_mean = np.mean(X[label_initial==0],axis=0)
c2_mean = np.mean(X[label_initial==1],axis=0)
centroids_updated = np.array([c1_mean,c2_mean])
print(centroids_updated)
plt.scatter(X[label_initial==0,0],X[label_initial==0,1],s=100,color='lightcoral',label='cluster1')
plt.scatter(X[label_initial==1,0],X[label_initial==1,1],s=100,color='skyblue',label='cluster2')
plt.scatter(centroid_inital[:,0], centroid_inital[:,1],s=100,color=['red','red'],marker='X',label='old centroids')
plt.scatter(centroids_updated[:,0], centroids_updated[:,1],s=100,color=['blue','blue'],marker='^',label='new centroids')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
kms = KMeans(n_clusters=2)
kms.fit(X)
```

```python
kms.labels_ # 군집을 1과 0 그룹으로 나눠서 보여줌 . 1그룹과 0그룹이 있는데 X에 있는 값이 어떤 그룹에 들어가있는지를 나타내줌
```

```python
kms.cluster_centers_
plt.scatter(X[:,0],X[:,1] , color = 'gray')
plt.scatter(kms.cluster_centers_[:,0], kms.cluster_centers_[:,1],color='red')
```
