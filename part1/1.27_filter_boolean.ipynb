```python
# 불린 인덱싱 (20세 미만의 조건을 충족하는 행을 필터링)
titanic_under20 = df[df['age'] < 20]
```

```python
#
titanic_under20 = df[df['age'] < 20]
titanic_under20 = titanic_under20.rename(columns={'sex':'gender'})
titanic_under20[titanic_under20['gender'] == 'female'].head(3)
```

```python
#age < 20 그리고 gender == 'female
mask = (df['age'] < 20) & (df['gender'] == 'female') #and 가 아닌 & 사용해야함  /  요고할때 요소 갯수가 맞아야함 / mask 함수로 묶어즘
df[mask]
```

```python
# 불린 인덱싱 (NOT 논리연산)
df.loc[~(df['age'] < 20)].head(5)
```

```python
# 나이가 10대(10~19세)인 승객 (AND 논리연산)
mask1 = (df.age >= 10) & (df.age < 20)
df_teenage = df[mask1]
df_teenage.head()
```

```python
# 나이가 10세 미만(0~9세)이고 여성인 승객 (AND 논리연산, loc 인덱서)
mask2 = (df.age < 10) & (df.gender == 'female')
df_female_under10 = df.loc[mask2]
df_female_under10.head()
```

```python
# 나이가 10세 미만(0~9세) 또는 60세 이상인 승객 (OR 논리연산)
# age, sex, alone 열만 선택 (loc 인덱서, fancy indexing)
mask3 = (df.age < 10) | (df.age >= 60)
df_under10_morethan60 = df.loc[mask3, ['age', 'gender', 'alone']]
df_under10_morethan60.head()
```

```python
# 불린 시리즈를 새로운 열로 추가
df['age_under_20'] = df['age'] < 20
df.head()
```

```python
# 1등석 , 2등석 , 3등석 pclass
# 선실별 승객수 len(데이터프레임)
df['pclass'].value_counts()
p1 = len(df[df['pclass'] == 1])
p2 = len(df[df['pclass'] == 2])
p3 = len(df[df['pclass'] == 3])
print(f'{'1등석'}:{p1}\n{'2등석'}:{p2}\n{'3등석'}:{p3}')
print("전체 승객 수:", len(df))
```
