```python
import seaborn as sns
import pandas as pd
```

```python
mpg = sns.load_dataset('mpg')
mpg.head(5)
```

```python
#groupby - cylinders별 마력의 평균
mpg.groupby('cylinders')['horsepower'].mean()
    # name 별 연식의 평균
mpg.groupby('name')['model_year'].mean()
    # name 이 공백을 기준을 제조사[0] - 모델[나머지]
    # 제조사 컬럼을 만들고 제조사를 추가
mpg['company'] = mpg['name'].apply(lambda x:x.split()[0])
mpg.head()
# iloc , loc 원하는 row col 데이터 추출 - mpg 15 이상인 모델
# high_mpg = mpg.loc['mpg'] >= 15
# high_mpg
# index를 재구성 - origin
# iloc , loc 이용하여 데이터를 row 와 col 별로 다양하게 출력

mpg.groupby('company')['name'].count()
mpg.set_index('company')
mpg.loc[:,['name','mpg']]
```
