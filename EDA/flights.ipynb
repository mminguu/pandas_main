```python
import seaborn as sns

flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.info() # category -> 문자형 데이터 속성도 category로 지정해주면 됌
```

```python
# year	month	passengers 를 가지고 피벗 연습하기 좋음
```

```python
flights.T # 행은 year	month	passengers  /  열은 연도별 값.
```

```python
flights.pivot(index='year', columns='month', values='passengers') # Pandas의 pivot()은 인자를 이름 붙여서 줘야 해
# index → 행으로 쓸 열  /  columns → 열(컬럼)으로 쓸 열  /  values → 안에 채울 값
```

```python
# 연도별 승객의 수 변화 lineplot  -  sns
# sns 시각화는 신뢰구간도 같이 표시하는 경우임.
# 신뢰구간을 없애고 싶다면 errorbar = None을 주면 됨
# errorbar = ('ci' , 95) 요로케 줘도 됌
import matplotlib.pyplot as plt
import seaborn as sns
# sns.lineplot(data = flights , x = 'year' , y = 'passengers' , errorbar = None)
sns.lineplot(data = flights , x = 'year' , y = 'passengers')
sns.lineplot(data = flights , x = 'year' , y = 'passengers',hue = 'month') # 월별로 구간

# 데이터가 변동될 수 있는 구간 ( 성수기 , 비수기 )
# 신뢰구간이 넓으면 불확실성이 큼 ( 데이터 분산이 크거나 심플수가 작은 경우)
# 신뢰구간이 좁으면 불확실성이 작음 ( 데이터 분산이 작거나 심플수가 많은 경우)

plt.legend(loc='best')
plt.show()



```

```python
# 년도별 월별 승객의 변화 시각화
sns.lineplot(data = flights, x = 'year' , y = 'passengers', hue = 'month')
```
